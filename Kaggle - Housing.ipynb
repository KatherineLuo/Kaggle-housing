```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
house_df = pd.read_csv('Kaggle Housing/train.csv')
test_df = pd.read_csv('Kaggle Housing/test.csv')
house_df.head()
house_price = house_df['SalePrice']
```

```python
# drop useless features

house_df.drop(['Id', 'Alley', 'Fence', 'PoolQC', 'MiscFeature'], axis=1, inplace = True)
```

```python
# MSSubClass
#average_price = house_df[["MSSubClass", "SalePrice"]].groupby(['MSSubClass'],as_index=False).mean()
sns.barplot(x='MSSubClass', y='SalePrice', data=house_df)
```

```python
# MSZoning
house_df['MSZoning'].unique()
sns.barplot(x='MSZoning', y='SalePrice', data=house_df)
```
